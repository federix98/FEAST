```python
import pandas as pd
from statistics import mean
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,7)
import pstats
import itertools
import numpy as np
import math
#from psutil import cpu_percent
import os
import csv
from statistics import mean
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,7)

def equation_filter(data, threshold = 0.04, penality = 1):
    #print(data[:10])
    _result = []
    for val in data:
        if val >= threshold:
            _result.append((threshold - val))
        else:
            _result.append((threshold - val))
    #print(_result[:10])
    return _result

def cumulate_data(data):
    _result, act_sum = [], 0
    for idx, val in enumerate(data):
        if idx == 0:
            _result.append(data[0])
            act_sum = data[0]
        else:
            _result.append(act_sum + val)
            act_sum += val
    return _result
    
def plot_data(title, xlabel, ylabel, data):
    colors = ['r', 'g', 'b', 'black', 'c', 'm']
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    for idx, dat in enumerate(data):
        plt.plot(dat[1], label=dat[0], c=colors[idx % len(colors)])
    plt.legend()
    plt.show()
    
    print("*********** PERFORMANCE ANALYSIS RESULTS ************")
    print("----------- " + title + " -----------------")
    for idx, dat in enumerate(data):
        print(dat[0], "avg", round(mean(dat[1]), 4), "s")
    print("-----------------------------------------------------\n\n")
            
def generate_data(component_layered_val, component_bb_val, component_ed_val):
    return [
        ["Layered Implementation", component_layered_val],
        ["Blackboard Implementation", component_bb_val],
        ["Event Driven Implementation", component_ed_val]
    ],[
        ["Layered Implementation", equation_filter(component_layered_val)],
        ["Blackboard Implementation", equation_filter(component_bb_val)],
        ["Event Driven Implementation", equation_filter(component_ed_val)]
    ],[
        ["Layered Implementation filtered", cumulate_data(equation_filter(component_layered_val))],
        ["Blackboard Implementation filtered", cumulate_data(equation_filter(component_bb_val))],
        ["Event Driven Implementation filtered", cumulate_data(equation_filter(component_ed_val))]
    ],[
        ["Layered Implementation", cumulate_data(component_layered_val)],
        ["Blackboard Implementation", cumulate_data(component_bb_val)],
        ["Event Driven Implementation", cumulate_data(component_ed_val)]
    ]


ED_Folder = "EventDriven/perf_data"
BB_Folder = "BB/perf_data"
LI_Folder = "Layered/perf_data"

ress = {}

MONITOR_LAYERED_RESPTIME = pd.read_csv(LI_Folder + '/MonitorRT_.txt', delimiter=";", header=None).values[0][:-1]
MONITOR_LAYERED_CPUTIME = pd.read_csv(LI_Folder + '/MonitorCPU_.txt', delimiter=";", header=None).values[0][:-1]
MONITOR_BB_RESPTIME = pd.read_csv(BB_Folder + '/MonitorRT_.txt', delimiter=";", header=None).values[0][:-1]
MONITOR_BB_CPUTIME = pd.read_csv(BB_Folder + '/MonitorCPU_.txt', delimiter=";", header=None).values[0][:-1]
MONITOR_ED_RESPTIME = pd.read_csv(ED_Folder + '/MonitorRT_.txt', delimiter=";", header=None).values[0][:-1]
MONITOR_ED_CPUTIME = pd.read_csv(ED_Folder + '/MonitorCPU_.txt', delimiter=";", header=None).values[0][:-1]

ANALYZER_LAYERED_RESPTIME = pd.read_csv(LI_Folder + '/AnalyzerRT_.txt', delimiter=";", header=None).values[0][:-1]
ANALYZER_LAYERED_CPUTIME = pd.read_csv(LI_Folder + '/AnalyzerCPU_.txt', delimiter=";", header=None).values[0][:-1]
ANALYZER_BB_RESPTIME = pd.read_csv(BB_Folder + '/AnalyzerRT_.txt', delimiter=";", header=None).values[0][:-1]
ANALYZER_BB_CPUTIME = pd.read_csv(BB_Folder + '/AnalyzerCPU_.txt', delimiter=";", header=None).values[0][:-1]
ANALYZER_ED_RESPTIME = pd.read_csv(ED_Folder + '/AnalyzerRT_.txt', delimiter=";", header=None).values[0][:-1]
ANALYZER_ED_CPUTIME = pd.read_csv(ED_Folder + '/AnalyzerCPU_.txt', delimiter=";", header=None).values[0][:-1]

PLANNER_LAYERED_RESPTIME = pd.read_csv(LI_Folder + '/PlannerRT_.txt', delimiter=";", header=None).values[0][:-1]
PLANNER_LAYERED_CPUTIME = pd.read_csv(LI_Folder + '/PlannerCPU_.txt', delimiter=";", header=None).values[0][:-1]
PLANNER_BB_RESPTIME = pd.read_csv(BB_Folder + '/PlannerRT_.txt', delimiter=";", header=None).values[0][:-1]
PLANNER_BB_CPUTIME = pd.read_csv(BB_Folder + '/PlannerCPU_.txt', delimiter=";", header=None).values[0][:-1]
PLANNER_ED_RESPTIME = pd.read_csv(ED_Folder + '/PlannerRT_.txt', delimiter=";", header=None).values[0][:-1]
PLANNER_ED_CPUTIME = pd.read_csv(ED_Folder + '/PlannerCPU_.txt', delimiter=";", header=None).values[0][:-1]

EXECUTOR_LAYERED_RESPTIME = pd.read_csv(LI_Folder + '/ExecutorRT_.txt', delimiter=";", header=None).values[0][:-1]
EXECUTOR_BB_RESPTIME = pd.read_csv(BB_Folder + '/ExecutorRT_.txt', delimiter=";", header=None).values[0][:-1]
EXECUTOR_ED_RESPTIME = pd.read_csv(ED_Folder + '/ExecutorRT_.txt', delimiter=";", header=None).values[0][:-1]
EXECUTOR_LAYERED_CPUTIME = pd.read_csv(LI_Folder + '/ExecutorCPU_.txt', delimiter=";", header=None).values[0][:-1]
EXECUTOR_BB_CPUTIME = pd.read_csv(BB_Folder + '/ExecutorCPU_.txt', delimiter=";", header=None).values[0][:-1]
EXECUTOR_ED_CPUTIME = pd.read_csv(ED_Folder + '/ExecutorCPU_.txt', delimiter=";", header=None).values[0][:-1]

# MONITOR ANALYSIS
MONITOR_DATA = [
    ["Layered Implementation", MONITOR_LAYERED_RESPTIME],
    ["Blackboard Implementation", MONITOR_BB_RESPTIME],
    ["Event Driven Implementation", MONITOR_ED_RESPTIME]
]

MONITOR_DATA, NEW_MONITOR_DATA, MONITOR_CUMULATIVE_DATA_1, MONITOR_CUMULATIVE_DATA_2 = generate_data(MONITOR_LAYERED_RESPTIME, MONITOR_BB_RESPTIME, MONITOR_ED_RESPTIME)

#plot_data("Monitor Response Time", 'Execution', 'Response Time', MONITOR_DATA)
#plot_data("Monitor Response Time", 'Execution', 'Response Time', NEW_MONITOR_DATA)
#plot_data("Monitor Response Time", 'Execution', 'Response Time', MONITOR_CUMULATIVE_DATA_2 + MONITOR_CUMULATIVE_DATA_1)
plot_data("Monitor Response Time", 'Execution', 'Response Time', MONITOR_CUMULATIVE_DATA_1)

# ANALYZER ANALYSIS
ANALYZER_DATA = [
    ["Layered Implementation", ANALYZER_LAYERED_RESPTIME],
    ["Blackboard Implementation", ANALYZER_BB_RESPTIME],
    ["Event Driven Implementation", ANALYZER_ED_RESPTIME]
]

ANALYZER_DATA, NEW_ANALYZER_DATA, ANALYZER_CUMULATIVE_DATA_1, ANALYZER_CUMULATIVE_DATA_2 = generate_data(ANALYZER_LAYERED_RESPTIME, ANALYZER_BB_RESPTIME, ANALYZER_ED_RESPTIME)

#plot_data("ANALYZER Response Time", 'Execution', 'Response Time', ANALYZER_DATA)
#plot_data("ANALYZER Response Time", 'Execution', 'Response Time', NEW_ANALYZER_DATA)
#plot_data("ANALYZER Response Time", 'Execution', 'Response Time', ANALYZER_CUMULATIVE_DATA_2 + ANALYZER_CUMULATIVE_DATA_1)
plot_data("ANALYZER Response Time", 'Execution', 'Response Time', ANALYZER_CUMULATIVE_DATA_1)

# PLANNER ANALYSIS
PLANNER_DATA = [
    ["Layered Implementation", PLANNER_LAYERED_RESPTIME],
    ["Blackboard Implementation", PLANNER_BB_RESPTIME],
    ["Event Driven Implementation", PLANNER_ED_RESPTIME]
]

PLANNER_DATA, NEW_PLANNER_DATA, PLANNER_CUMULATIVE_DATA_1, PLANNER_CUMULATIVE_DATA_2 = generate_data(PLANNER_LAYERED_RESPTIME, PLANNER_BB_RESPTIME, PLANNER_ED_RESPTIME)

#plot_data("PLANNER Response Time", 'Execution', 'Response Time', PLANNER_DATA)
#plot_data("PLANNER Response Time", 'Execution', 'Response Time', NEW_PLANNER_DATA)
#plot_data("PLANNER Response Time", 'Execution', 'Response Time', PLANNER_CUMULATIVE_DATA_2 + PLANNER_CUMULATIVE_DATA_1)
plot_data("PLANNER Response Time", 'Execution', 'Response Time', PLANNER_CUMULATIVE_DATA_1)

# EXECUTOR ANALYSIS
EXECUTOR_DATA = [
    ["Layered Implementation", EXECUTOR_LAYERED_RESPTIME],
    ["Blackboard Implementation", EXECUTOR_BB_RESPTIME],
    ["Event Driven Implementation", EXECUTOR_ED_RESPTIME]
]

EXECUTOR_DATA, NEW_EXECUTOR_DATA, EXECUTOR_CUMULATIVE_DATA_1, EXECUTOR_CUMULATIVE_DATA_2 = generate_data(EXECUTOR_LAYERED_RESPTIME, EXECUTOR_BB_RESPTIME, EXECUTOR_ED_RESPTIME)

#plot_data("EXECUTOR Response Time", 'Execution', 'Response Time', EXECUTOR_DATA)
#plot_data("EXECUTOR Response Time", 'Execution', 'Response Time', NEW_EXECUTOR_DATA)
plot_data("EXECUTOR Response Time", 'Execution', 'Response Time', EXECUTOR_CUMULATIVE_DATA_1)
#plot_data("EXECUTOR Response Time", 'Execution', 'Response Time', CUMULATIVE_DATA_1)




#plot_data("Monitor Response Time", 'Execution', 'Response Time', MONITOR_DATA)#MONITOR_CUMULATIVE_DATA_1)
#plot_data("Monitor Response Time", 'Execution', 'Response Time', NEW_MONITOR_DATA)
#plot_data("Monitor Response Time", 'Execution', 'Response Time', MONITOR_CUMULATIVE_DATA_1)
#plot_data("ANALYZER Response Time", 'Execution', 'Response Time', ANALYZER_CUMULATIVE_DATA_1[:100])
#plot_data("PLANNER Response Time", 'Execution', 'Response Time', PLANNER_CUMULATIVE_DATA_1)
#plot_data("EXECUTOR Response Time", 'Execution', 'Response Time', EXECUTOR_CUMULATIVE_DATA_1)
```
